```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
global_aus_path = "Global_EV_sales(2020-2022).csv"


global_aus_df = pd.read_csv(global_aus_path)
global_aus_df
```

```python
aus_evsales_path = "Annual_Vechile_sale.csv"


aus_evsales= pd.read_csv(aus_evsales_path)
aus_evsales
```

```python
time = aus_evsales['Year']
costs = aus_evsales['Total Sales']

plt.plot(time, costs)
plt.xlabel("Year")
plt.ylabel("EV sales")
plt.title("Electric car sales in Australia")
plt.show()
```

```python
time = global_aus_df['Month']
costs = global_aus_df['EV sales']

plt.plot(time, costs)
plt.xlabel("Year")

plt.ylabel("EV sales")
plt.title("Global and aus ev sales")
plt.show()
```
