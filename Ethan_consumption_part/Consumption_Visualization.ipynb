```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

merged_consumption_df_path = "my_data/cleaned_data/merged_consumption.csv"
merged_consumption_df = pd.read_csv(merged_consumption_df_path)

```

```python
#Line Graph Comparation
electricity_df = merged_consumption_df[merged_consumption_df['Fuel_usage'] == 'Electricity']
gasoline_df = merged_consumption_df[merged_consumption_df['Fuel_usage'] == 'Gasoline']

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(electricity_df['Models'], electricity_df['City_FE'],  'g-', label='City_FE_Electricity')
ax.plot(electricity_df['Models'], electricity_df['Hwy_FE'], 'g--', label='Hwy_FE_Electricity')
ax.plot(electricity_df['Models'], electricity_df['Comb_FE'], 'g:', label='Comb_FE_Electricity')
ax.legend()

plt.xticks(rotation=90, fontsize=6)

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(gasoline_df['Models'], gasoline_df['City_FE'], 'r-', label='City_FE_Gasoline')
ax.plot(gasoline_df['Models'], gasoline_df['Hwy_FE'], 'r--', label='Hwy_FE_Gasoline')
ax.plot(gasoline_df['Models'], gasoline_df['Comb_FE'], 'r:', label='Comb_FE_Gasoline')
ax.legend()


plt.xticks(rotation=90, fontsize=6)

plt.show()

```

```python
# Extracting electricity and gasoline data
electricity_df = merged_consumption_df[merged_consumption_df["Fuel_usage"]=="Electricity"]
gasoline_df = merged_consumption_df[merged_consumption_df["Fuel_usage"]=="Gasoline"]

# Plotting City_FE for electricity and gasoline in separate scatter plots
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(electricity_df["Models"], electricity_df["City_FE"], color="g", label="City_FE (Electricity)")
ax.legend()

plt.xticks(rotation=90, fontsize=6)

fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(gasoline_df["Models"], gasoline_df["City_FE"], color="r", label="City_FE (Gasoline)")
ax.legend()

plt.xticks(rotation=90, fontsize=6)

# Plotting Hwy_FE for electricity and gasoline in separate scatter plots
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(electricity_df["Models"], electricity_df["Hwy_FE"], color="g", label="Hwy_FE (Electricity)")
ax.legend()

plt.xticks(rotation=90, fontsize=6)

fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(gasoline_df["Models"], gasoline_df["Hwy_FE"], color="r", label="Hwy_FE (Gasoline)")
ax.legend()

plt.xticks(rotation=90, fontsize=6)

# Plotting Comb_FE for electricity and gasoline in separate scatter plots
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(electricity_df["Models"], electricity_df["Comb_FE"], color="g", label="Electricity_Comb_Fuel_Efficiency")
ax.legend()

plt.xticks(rotation=90, fontsize=6)

fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(gasoline_df["Models"], gasoline_df["Comb_FE"], color="r", label="Gasoline_Comb_Fuel_Efficiency")
ax.legend()

plt.xticks(rotation=90, fontsize=6)
```

```python
# Box Plot for Electricity and Gasoline
electricity_data = merged_consumption_df[merged_consumption_df["Fuel_usage"] == "Electricity"]["Comb_FE"]
gasoline_data = merged_consumption_df[merged_consumption_df["Fuel_usage"] == "Gasoline"]["Comb_FE"]

fig, ax = plt.subplots()
ax.boxplot([electricity_data, gasoline_data], labels=["Electricity", "Gasoline"])
ax.set_ylabel("Fuel_Efficiency(MPG)")
ax.set_title("Comb_Fuel_Efficiency for Electricity and Gasoline")
plt.legend(title="Unit(Miles/Gallon))")
plt.show()
```
