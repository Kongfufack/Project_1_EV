```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

EV_consumption_path = "my_data/EV_consumption_data.csv"
EV_consumption = pd.read_csv(EV_consumption_path)

Tradition_consumption_path = "my_data/Tradition_consumption_data.csv"
Tradition_consumption = pd.read_csv(Tradition_consumption_path,encoding='ISO-8859-1')
```

```python
EV_consumption.head()
```

```python
#clean EV_consumption
EV_consumption = EV_consumption.copy()
EV_consumption = EV_consumption.dropna(subset=['Model Year'])
EV_consumption.insert(0, 'Models', '')
EV_consumption['Models'] = EV_consumption['Model Year'].astype(str) + '_' + EV_consumption['Mfr Name'] + '_' + EV_consumption['Division'] + '_' + EV_consumption['Carline']
EV_consumption = EV_consumption[EV_consumption['Fuel Unit - Conventional Fuel'] != 'KW-HR/100Miles']
EV_consumption.drop(columns=['Model Year','Mfr Name','Division','Carline','FE Rating (1-10 rating on Label)',"Range1 - Model Type Driving Range - Conventional Fuel","City Unadj FE - Conventional Fuel","Hwy Unadj FE - Conventional Fuel","Comb Unadj FE - Conventional Fuel"], inplace=True)
EV_consumption.head()
EV_consumption_cleaned = EV_consumption
```

```python
#EV_consumption_cleaned.to_csv("my_data/cleaned_data/EV_consumption_cleaned.csv", index=False)
```

```python
EV_consumption_cleaned.head()
```

```python
#Clean Tradition_consumption
Tradition_consumption = Tradition_consumption.copy()
Tradition_consumption.insert(0, 'Models', '')
Tradition_consumption['Models'] = Tradition_consumption['Model Year'].astype(str) + '_' + Tradition_consumption['Division'] + '_' + Tradition_consumption['Carline'] 
Tradition_consumption.drop(columns=['Model Year','Division','Carline',"City Unadj FE - Conventional Fuel","Hwy Unadj FE - Conventional Fuel","Comb Unadj FE - Conventional Fuel"], inplace=True)
Tradition_consumption.head()
```

```python
Tradition_consumption_cleaned = Tradition_consumption

```

```python
Tradition_consumption_cleaned.to_csv("my_data/cleaned_data/Tradition_consumption_cleaned.csv", index=False)
```

```python
EV_consumption_cleaned = EV_consumption_cleaned.rename(columns={
    'Fuel Unit - Conventional Fuel': 'Fuel_unit',
    'Fuel Usage Desc - Conventional Fuel': 'Fuel_usage',
    'City FE (Guide) - Conventional Fuel': 'City_FE',
    'Hwy FE (Guide) - Conventional Fuel': 'Hwy_FE',
    'Comb FE (Guide) - Conventional Fuel': 'Comb_FE'
})
cols = EV_consumption_cleaned.columns.tolist()
cols = cols[:1] + [cols[2]] + [cols[1]] + cols[3:]
EV_consumption_cleaned = EV_consumption_cleaned[cols]



EV_consumption_cleaned.head()
```

```python
Tradition_consumption_cleaned = Tradition_consumption_cleaned.rename(columns={
    'Fuel Unit - Conventional Fuel': 'Fuel_unit',
    'Fuel Usage Desc - Conventional Fuel': 'Fuel_usage',
    'City FE (Guide) - Conventional Fuel': 'City_FE',
    'Hwy FE (Guide) - Conventional Fuel': 'Hwy_FE',
    'Comb FE (Guide) - Conventional Fuel': 'Comb_FE'
})
Tradition_consumption_cleaned = Tradition_consumption_cleaned.drop(['Fuel Unit Desc - Conventional Fuel'], axis=1)
Tradition_consumption_cleaned[['City_FE', 'Hwy_FE', 'Comb_FE']] = Tradition_consumption_cleaned[['City_FE', 'Hwy_FE', 'Comb_FE']].astype(float)



Tradition_consumption_cleaned.head()
```

```python
#Merge
merged_consumption_df = pd.concat([EV_consumption_cleaned, Tradition_consumption_cleaned], axis=0, ignore_index=True)

merged_consumption_df.head()
```

```python
merged_consumption_df['Fuel_usage'] = merged_consumption_df['Fuel_usage'].apply(lambda x: 'Gasoline' if x != 'Electricity' else x)
merged_consumption_df.head()
```

```python
merged_consumption_df.to_csv("my_data/cleaned_data/merged_consumption.csv", index=False)
```
